```python
import pandas as pd
```

```python
dataset = pd.read_csv(
        filepath_or_buffer = "../dataset/Dataset.csv",
        header = 11,
        skiprows = [12],
        index_col = False,
        low_memory = False)
```

```python
dataset = dataset[["ID", "Trading date", "Trading time", "SecType", "Last"]]
dataset = dataset.rename(columns = {"Trading date" : "TradingDate", "Trading time" : "TradingTime"})

dataset
```

```python
dataset = dataset[dataset["TradingTime"] != "00:00:00.000"]
dataset = dataset[dataset["TradingDate"].notnull()]
dataset
```

```python
dropedDataset = dataset.drop_duplicates()
dropedDataset
```

```python
values = dataset[
    (dataset["ID"] == "ABEO.FR") & 
    (dataset["TradingDate"] == "08-11-2021") &
    (dataset["TradingTime"].str.startswith("11:44:53.551"))]

values
```

```python
cleanDataset = dataset[["ID", "SecType", "TradingDate", "TradingTime", "Last"]].drop_duplicates()
cleanDataset
```

```python
cleanDataset["TradingHour"] = cleanDataset["TradingTime"].str[0:2]
cleanDataset
```

```python
response = cleanDataset[(cleanDataset["SecType"] == "E") & (cleanDataset["ID"].str.endswith(".FR"))]
response = response.groupby(["TradingDate", "TradingHour", "ID"]).mean("Last")
response
```
